## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

The type of my lane is ranking/scoring. This is because I will be calculating an 
opportunity score for each page and ranking them by it, not sorting pages into 
fixed categories or grouping similar pages together.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

I would be predicting the "opportunity" of a page — the gap between its expected 
CTR and its actual CTR. Expected CTR is not something that already exists in the 
data; I will define it myself, for example by taking the average CTR of pages at 
a similar position. A bigger gap means the page ranks well but is not getting the 
clicks it should, so it is worth fixing. This comes from a defined rule I create, 
not an outcome that was already observed.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

My success metric would be precision@K. After ranking pages by opportunity score, 
I would check the top K pages (for example, the top 20) and see how many are 
actually real opportunities, pages that rank well but are genuinely underperforming 
on clicks. If most of the top K pages turn out to be real opportunities, the ranking 
is working. This metric fits because the output is a short list someone will act 
on, so what matters most is whether the top of the list is trustworthy, not whether 
every single page in the dataset is scored perfectly.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A simple rule (if-statement) like "if CTR < 2%, flag it" will not work here, because 
the same CTR number can mean different things depending on the page's position. For 
example, a page at position 2 with 2% CTR is genuinely underperforming, since top 
positions normally get much higher clicks. But a page at position 9 with the same 
2% CTR might be completely normal, since lower positions naturally get fewer clicks. 
A fixed rule cannot tell these two cases apart, because it only looks at CTR and 
ignores position. My opportunity score fixes this by comparing each page's CTR to 
the expected CTR for its own position, so the same 2% means something different 
depending on where the page ranks.
