For the SEO team, deciding which pages to improve first, we will build a ranked priority

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv("C:/Users/tayyaba/Desktop/Flyrank-ML-Internship/data/raw/content_refresh_anonymized.csv")

# avg_position = 0 means "no data", so remove those rows
has_position = df[df["avg_position"] > 0]

# look only at pages ranking in the top 10
top10 = has_position[has_position["avg_position"] <= 10]

# find the middle CTR value for top 10 pages
median_ctr = top10["ctr"].median()

# pages that rank well but have below-average clicks
low_ctr_pages = top10[top10["ctr"] < median_ctr]

print("Total pages:", len(df))
print("Pages ranking top 10:", len(top10))
print("Of those, below-median CTR:", len(low_ctr_pages))
print("Percent underperforming:", round(len(low_ctr_pages) / len(top10) * 100, 1), "%")
```

Out of 30,000 pages, 12,983 rank in the top 10 positions. Of those, 6,432 pages 
(49.5%) get fewer clicks than the median, even though they rank well. This shows 
a real gap worth fixing — good ranking position does not always mean good clicks, 
and there are enough cases here to make this lane worth building on.
